```python
#Adding dataframes for 18-22 to one dataframe as a training set
stat_train = pd.concat([stat22, stat21, stat20, stat19, premStat22, premStat21, premStat20, premStat19, serStat22, serStat21, serStat20, serStat19], ignore_index=True)
```

```python
stat_test = stat_test.drop(columns=['W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts/MP'])
stat_train = stat_train.drop(columns=['W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts/MP'])
```

```python
stat_train
```

```python
#Importing necessary methods for model building and evaluation
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Creating response and independent variable sets
x_train, x_test = stat_train.drop(columns=['Squad', 'Pts', 'Rk']), stat_test.drop(columns=['Squad', 'Pts', 'Rk'])
y_train, y_test = stat_train['Pts'], stat_test['Pts']
```

```python
#Creating the model
rf = RandomForestRegressor(n_estimators=300, random_state=10)

#Training the model
rf.fit(x_train, y_train)
```

```python
#Making predictions and evaluating the model
y_pred = rf.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

```python
#Investigating the important variables
importances = rf.feature_importances_

feature_names = x_train.columns
plt.barh(feature_names, importances)
plt.xlabel('Feature Importance')
plt.ylabel('Feature Name')
plt.title('Feature Importance Plot')
plt.show()
```
